经过前面的讨论，我们已经可以取出特定的数据对象了，接下来我们要实现这些特定数据对象之间的运算。

````r
# 首先，我们讨论就是向量的运算
# 向量加
x <- 1:4
y <- rep(1, 4)
z <- x + y
print(z)
# 数乘
a <- 10
print(a * x)
y <- c(2, 4, 6, 8, 10)
# 函数运算时，将对向量中的每个元素逐一求值
print(exp(y))
print(sqrt(y))
print(log(y))
````

````r
# 向量可以看作是一个存储了一系列元素的表，那么就有表上的运算了
x <- c(10, 6, 4, 7, 8)
# 求最大最小，范围
print(min(x))
print(max(x))
print(range(x))
# 求最大最小的索引
print(which.max(x))
print(which.min(x))
# 求所有元素的和，积以及求长度
print(sum(x))
print(prod(x))
print(length(x))
````

向量上的运算我们介绍完了，接着就是考虑在矩阵上的运算了。

````r
# 生成单位矩阵的方法
B <- diag(2)
print(B)
v <- rep(1, 5)
print(v)
I5 <- diag(v)
print(I5)
# 提取矩阵的对角线元素
A <- matrix(1:9, nrow = 3)
print(A)
print(diag(A))
# 连续使用两次即可生成对应元素的对角矩阵
print(diag(diag(A)))
````

上面介绍了矩阵中提取对角线元素和构造对角矩阵的方法，下面主要介绍矩阵之间的运算。

````r
A <- matrix(1:4, 2, 2)
B <- diag(2)

print(A + B)
print(A - B)
# 直接使用乘除运算符只是单纯对其中元素逐一运算，而非矩阵上的乘除运算
print(A * B)
print(B / A)
# 矩阵意义上的乘
print(A %*% B)
````

除了矩阵之间的加减乘除，还有矩阵本身的各种运算，比如求特征值，奇异值分解，求行列式等

````r
A <- matrix(1:4, 2, 2)
B <- matrix(rep(1, 4), 2, 2)
# 计算 A 和 B 的 kronecker 积
# 将 A 中的每个元素乘上矩阵 B ，将结果按照 A 矩阵的排列方式排列
print(kronecker(A, B))
# 求矩阵的维数
print(dim(A))
print(nrow(A))
print(ncol(A))
# 求矩阵转置
print(t(A))
# 求行列式
print(det(A))
# 求特征值和特征向量
print(eigen(A))
# 求矩阵的逆
print(solve(A))
````

矩阵上和矩阵之间的基本运算我们已经介绍完毕了，矩阵还有另一个大用处就是用来求解线性方程组。

$$ AX = B \rightarrow X = A^{-1}B$$

````r
A <- matrix(1:9, nrow = 3, byrow = T)
A[3, 3] <- 10
b <- rep(1, 3)
x <- solve(A, b)
D <- solve(A)
print(x)
print(D)
````

这样来求解线性方程组，我们要确保系数矩阵是可逆的，判断是否可逆最简单的方法就是求解其行列式值是否等于0，但是计算机在与0作比较时会带来新的问题，先看下面的这个例子。

````r
B <- diag(c(100000000000, 0.00001))
print(B)
det(B)
solve(B)
````

显然，这个矩阵 $B$ 的行列式值是不为0的，但是在求解该矩阵的逆的时候报错了。这是因为虽然该矩阵在数学上可逆，但是在数值计算上是病态的，数值相差接近 $10^{16}$ 级，几乎接近双精度浮点数的边界，所以程序为防止溢出的意外错误选择报错。

为了解决这种问题，我们需要在使用 ```solve()``` 之前判断矩阵是否在数值计算上可逆。

````r
# rcond() 计算矩阵的倒条件数，将其与机器的双精度浮点数边界比较
rcond(B) > .Machine$double.eps
if (rcond(B) > .Machine$double.eps) {
    B_inv <- solve(B)
}
if (rcond(B) <= .Machine$double.eps) {
    cat("the matrix is computationally sigular!")
}
````

下面我们对矩阵做一些操作，比如拼接两个矩阵，修改矩阵的某一部分，将其变为三角矩阵。

````r
A <- matrix(1:9, 3, 3)
print(A)
B <- matrix(rep(1, 9), 3, 3)
print(B)
print(cbind(A, B))
print(rbind(A, B))

print(lower.tri(A, diag = T))
B[lower.tri(A, diag = T)] <- 0
print(B)
````

矩阵上的运算还有各种各样的分解方式，例如特征根分解，奇异值分解等等，R 中也是内置了这些分解的函数。

````r
# 奇异值分解
A <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 2, byrow = TRUE)
# 需要注意的是，奇异值分解得到的不是奇异矩阵，而是奇异值构成的向量
s <- svd(A)
cat("U=\n")
print(s$u)
cat("d=\n")
print(s$d)
cat("V=\n")
print(s$v)

cat("还原A：\n")
print(s$u %*% diag(s$d) %*% t(s$v))
````

````r
# Cholesky分解
B <- matrix(c(4, 2, 2, 3), nrow = 2, byrow = TRUE)

R <- chol(B)
cat("Cholesky分解得到的上三角矩阵R：\n")
print(R)

cat("t(R)%*%R=\n")
print(t(R) %*% R)
````

````r
# QR分解
C <- matrix(c(1, 2, 3, 4, 5, 6), nrow = 3, byrow = TRUE)

qrc <- qr(C)
Q <- qr.Q(qrc)
R <- qr.R(qrc)

cat("Q=\n")
print(Q)
cat("R=\n")
print(R)

cat("Q%*%R=\n")
print(Q %*% R)
````
