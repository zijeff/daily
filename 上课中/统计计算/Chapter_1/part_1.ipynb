这里主要是解释R中的一些数据类型，R虽然作为解释型语言，但是了解对应的数据类型可以帮助我们编程少出bug。但是为了更好的阅读体验，这里只涉及到了原子类型和一些特殊值，数据结构将会在后面涉及。

| 类别     | 名称         | 含义                     | 示例              | 说明 |
|----------|--------------|--------------------------|-------------------|------|
| 原子类型 | logical      | 逻辑型                   | TRUE, FALSE       | 表示真和假 |
| 原子类型 | integer      | 整数型                   | 1L, 2L            | 需加 `L` 才表示整数 |
| 原子类型 | numeric      | 数值型（双精度）         | 3.14, 10          | 默认数值类型，底层通常是 double |
| 原子类型 | character    | 字符型                   | "hello", "R"      | 字符串需要用引号括起来 |
| 原子类型 | complex      | 复数型                   | 2+3i              | 用于表示复数 |
| 原子类型 | raw          | 原始字节型               | charToRaw("A")    | 用于底层字节操作 |
| 特殊值   | NA           | 缺失值                   | NA                | 表示数据缺失 |
| 特殊值   | NULL         | 空对象                   | NULL              | 表示没有对象或空值 |
| 特殊值   | NaN          | 非数字                   | 0/0               | 表示非法数值运算结果 |
| 特殊值   | Inf          | 正无穷                   | 1/0               | 表示正无穷大 |
| 特殊值   | -Inf         | 负无穷                   | -1/0              | 表示负无穷大 |

````r
x <- 1
mode(x)     # 查看对象的类型
length(x)   # 查看对象的长度

A <- "Gomphotherium"; compar <- TRUE; z <- 1i
mode(A); mode(compar); mode(z)

N <- 2.1e23; print(N)
x <- 5/0; print(x)
# 取指数函数分析，无穷大减去无穷小，结果非数，也就是计算机不确定这是个啥。
exp(x); exp(-x); print(x-x) 
````

从上述运行结果可以看出，R语言中整数的底层类型实际上是双精度浮点数，而非预想的整型，要使其类型为整型，则需要我们人为指定，而这也就引入了接下来我们讨论的问题——不同数据类型之间的转换。

````r
z <- 0:9; print(z)
digits <- as.character(z); print(digits)
x <- as.numeric(digits); print(x)
mode(x)
````

类型之间的转换我们采用内置函数```as.()```，更多的用法可以自行查询，这里不过多赘述。接下来，我们主要讨论```getwd(), setwd("")```的用法与注意事项。

如果各位有计算机基础，理解起来是很容易的。因为我们在统计数据的时候，往往需要从文件中读取数据，再将结果数据保存到特定的文件中。那么为了告诉计算机去哪里找文件，我们就需要设定工作目录了，然后 R 语言就会自动地在工作目录里面寻找文件。

````r
# getwd()用来显示当前目录，默认为当前目录
getwd()
# 设定工作目录，这里运行将会显示报错，因为当前工作目录就在这里，所以是无法修改的
setwd("")
````

但是，我个人并不建议使用这两个代码。因为使用后会导致代码可移植性变差。换一台电脑或换个文件夹后，代码很可能就跑不通。相比直接写死```setwd()```，很多时候更推荐项目化管理，也就是在开始你的代码之前规划后文件设计，然后使用相对目录的写法。

关于 R 语言中读取数据和写入数据的问题，可以查看当前目录下的```read_and_write```文件夹。接下来我们主要讨论的是 R 语言中的一些数据结构，首先就是向量。

````r
#向量的生成
x <- c(1, 2, 3, 4, 5)
print(x)
y <- 1:5
print(y)
ch <- c("sa", "ba")
print(ch)
z <- seq(1, 5)
print(z)
````

如你所见，向量就是具有相同数据类型的元素的有序序列，那么我们在构造连续序列的时候可以选择一个一个手动输入。但是，在面对高维度的向量时，我们希望有更方便的构造方法，所以下面主要介绍```seq, rep``` 这两个函数的用法。

````r
w <- seq(from = 0, to = 10, by = 1)
print(w)
# 使用 length.out 将给定区间划分为间距相等的点
z <- seq(from = 0, to = 10, length.out = 10)
print(z)

print(rep(1, 3))
print(rep(c(1, 2), 2))
print(rep(1:3, 1:3))
print(rep(1:2, each = 2))
````

```gl()``` 是 R 语言中用来生成因子（factor levels）的函数，常用于构造分组变量。
它特别适合在实验设计、分组数据、重复处理水平时使用。

````r
gl(2, 3)
# 指定长度，长度不够循环补齐，够了就截断
gl(2, 3, length = 4)
# 自定义标签名
gl(2, 2, label = c("F", "M"))
````
