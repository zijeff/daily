R 语言作为一门统计的语言，其绘图功能也是十分强大的，所以该部分我们主要讨论 R 中的常见绘图方式。

首先，我们讨论柱形图的绘图。

```r
# 加载 vcd 包，并调用其中的 Arthritis 数据集
library(vcd)
print(head(Arthritis))
# 这个数据集是有关于关节炎治疗效果的数据
table1 <- table(Arthritis$Improved)
print(table1)

jpeg("./barplots/picture1.jpeg", height = 760, width = 800, quality = 100)
# 将画图区域划分为两行两列
opar <- par(mfrow = c(2, 2))
# 画三张图
barplot(table1, xlab = "Improvement", ylab = "Frequency") # 1. 默认的图，颜色为灰色
barplot(table1, xlab = "Improvement", ylab = "Frequency", col = 2:4) # 2. 赋予了三种颜色
barplot(table1, xlab = "Improvement", ylab = "Frequency", col = 3:5, horiz = TRUE) # 3. 改变绘图方向
# 恢复画图参数，便于后续操作
par(opar)
# 关闭画图设备
dev.off()
```

```r
table2 <- table(Arthritis$Improved, Arthritis$Treatment)
print(table2)

jpeg("./barplots/picture2.jpeg", height = 400, width = 1000, quality = 100)
# 使用一行两列的画图布局
opar <- par(mfrow = c(1, 2))
# legend = rownames(table2) 表明使用table2的行名
barplot(table2,
    main = "Stacked Barplot", xlab = "Treatment", ylab = "Frequency",
    col = c("red", "blue", "green"), legend = rownames(table2)
)
# beside = TRUE 表明柱形图并列排放，而非叠放
barplot(table2,
    main = "Grouped Barplot", xlab = "Treatment", ylab = "Frequency",
    col = c("red", "blue", "green"), legend = rownames(table2), beside = TRUE
)
par(opar)
dev.off()
```

```r
# 该部分与上面代码并无本质差别，只是单纯换了一下颜色
jpeg("./barplots/picture3.jpeg", height = 400, width = 1000, quality = 100)
opar <- par(mfrow = c(1, 2))
barplot(table2,
    main = "Stacked Barplot", xlab = "Treatment", ylab = "Frequency",
    col = c("coral1", "orange", "yellow"), legend = rownames(table2)
)
barplot(table2,
    main = "Grouped Barplot", xlab = "Treatment", ylab = "Frequency",
    col = c("blue1", "dodgerblue", "skyblue1"), legend = rownames(table2), beside = T
)
par(opar)
dev.off()
```

好的，柱形图的部分我们介绍完毕了。从这个简单的例子可以看出，R 的绘图功能还是比较好玩的，可以选择自己喜欢的颜色，如果不知道你想要的颜色名字用十六进制表示也是可以的。但是，不要搞一些反差太大的颜色，放一起不太好看😭。

接下来，我们讨论箱形图的绘制。

```r
print(head(mtcars))
mpg <- mtcars$mpg
print(mpg)

jpeg("./boxplots/mbox1.jpeg", height = 800, width = 800, quality = 100)
boxplot(mpg)
dev.off()

jpeg("./boxplots/mbox2.jpeg", height = 800, width = 800, quality = 100)
opar <- par(mfrow = c(2, 2))
boxplot(mpg ~ cyl,
        data = mtcars,
        main = "mtcars data", xlab = "cylinders", ylab = "miles per gallon")
boxplot(mpg ~ cyl,
        data = mtcars, col = 4,
        main = "mtcars data", xlab = "cylinders", ylab = "miles per gallon")
boxplot(mpg ~ cyl,
        data = mtcars, col = rainbow(3),
        main = "mtcars data", xlab = "cylinders", ylab = "miles per gallon")
boxplot(mpg ~ cyl,
        data = mtcars, col = heat.colors(3, alpha = 0.4),
        main = "mtcars data", xlab = "cylinders", ylab = "miles per gallon")
par(opar)
dev.off()
```

好的，箱形图我们也介绍完了，接下来讨论一下点图的绘制。

```r
jpeg("./dotchart/mdot1.jpeg", height = 800, width = 800, quality = 100)
dotchart(mtcars$mpg,
    labels = row.names(mtcars),
    main = "dotchart of mtcars data"
)
dev.off()

# 按照 mpg 排序得到新的数据
mtcars2 <- mtcars[order(mtcars$mpg), ]
# 将 cyl 转为因子类型，便于后面的分类
mtcars2$cyl <- factor(mtcars2$cyl)
mtcars2$color[mtcars2$cyl == 4] <- "red"
mtcars2$color[mtcars2$cyl == 6] <- "blue"
mtcars2$color[mtcars2$cyl == 8] <- "darkgreen"

jpeg("./dotchart/mdot2.jpeg", height = 800, width = 800, quality = 100)
dotchart(mtcars2$mpg,
    labels = row.names(mtcars2),
    groups = mtcars2$cyl, gcolor = "black",
    color = mtcars2$color, pch = 19,
    main = "dotchart of mtcars data"
)
dev.off()
```

然后，我们讨论的是直方图的绘制。

```r
mpg=mtcars$mpg

jpeg("./hist/picture.jpeg", height=800, width=960, quality = 100)
opar=par(mfrow=c(2,2))
# 这是最简单的直方图绘制，全部采用默认参数
hist(mpg)
# 数据范围划分为12个区间，自定义图名称和横轴名称
hist(mpg, breaks=12, col="grey", 
     xlab="Miles Per Gallon", main="histogram of mtcars")
# 采用freq参数，绘制概率密度直方图
hist(mpg, breaks=12, freq=F, col="darkseagreen", 
     xlab="Miles Per Gallon", main="histogram of mtcars")
# 叠加核密度曲线，更加直观看出数据分布情况
lines(density(mpg), col="darkgreen", lwd=2)

hist(mpg, breaks=12, freq=F, col="darkseagreen", 
     xlab="Miles Per Gallon", main="histogram of mtcars")
lines(density(mpg), col="darkgreen", lwd=2)
# 在上面的基础上再加入正态曲线的绘制
xfit=seq(min(mpg), max(mpg), length=40)
yfit=dnorm(xfit, mean(xfit), sd(xfit))
lines(xfit, yfit, col="tomato4", lwd=2)

par(opar)
dev.off()
```

饼状图的绘制，额啊，懒得想连接词了。

```r
favor <- read.csv("./pie/favoriate.csv", header = F)
print(head(favor))

table3 <- table(favor)
print(table3)

jpeg("./pie/favor.jpeg", height = 1200, width = 1200, quality = 100)
opar <- par(mfrow = c(2, 2))
# 默认参数绘制
pie(table3)

# 拼接出每种的名字
lab2 <- paste(names(table3), "\n", table3, sep = "")
pie(table3, labels = lab2, col = rainbow(length(table3)))

# 计算对应类别所占百分比，然后四舍五入
freq <- round(table3 / sum(table3) * 100)
lab3 <- paste(names(table3), "\n", freq, "%", sep = "")
pie(table3,
    labels = lab3,
    col = c("olivedrab2", "chartreuse3", "honeydew3", "goldenrod1", "peru")
)

# 与上一幅图不同点在于采用了热度色，并设置了不透明度
freq <- round(table3 / sum(table3) * 100)
lab3 <- paste(names(table3), "\n", freq, "%", sep = "")
pie(table3,
    labels = lab3,
    col = heat.colors(5, alpha = 0.4)
)
par(opar)
dev.off()
```

下面这部分代码我们主要讨论两个问题：一是绘图设备怎么用，二是怎么在一个窗口里显示多张图。

```r
## 绘图设备
mydata <- read.csv("./plot/dose.csv")
print(head(mydata))

attach(mydata)

# 直接在终端显示图片结果，可以注释掉
# plot(dose, drugA)
# plot(dose, drugA, type = "b")
# plot(dose, drugA, type = "b", lty = 4, pch = 21, lwd = 2, cex = 1.5, col = "red")

# R 语言支持多种图片格式
pdf("./plot/dose1.pdf")
plot(dose, drugA)
dev.off()

postscript("./plot/dose2.eps")
plot(dose, drugA, type = "b")
dev.off()

jpeg("./plot/dose3.jpeg", height = 600, width = 600, quality = 100)
plot(dose, drugA, type = "b", lty = 4, pch = 21, lwd = 2, cex = 1.5, col = "red")
dev.off()

# 列出当前绘图设备
dev.list()
dev.cur()
# 切换与关闭
dev.set(4)
dev.off(4)
dev.off()

# 多图显示
# 使用 split.screen ，关闭所有图形设备，清空图形环境
graphics.off()

# 划分布局
split.screen(c(2, 1))
split.screen(c(1, 3), screen = 2)
screen(1)
plot(dose, drugA, type = "b")
screen(3)
plot(dose, drugA, type = "l", lty = 1, lwd = 1, cex = 1, col = "red")
screen(4)
plot(dose, drugA, type = "l", lty = 2, lwd = 2, cex = 2, col = "blue")
screen(5)
plot(dose, drugA, type = "l", lty = 3, lwd = 3, cex = 3, col = "green")
close.screen(all = TRUE)

# layout
mypar <- par(no.readonly = TRUE) # 存储当前图形设备设置
layout(matrix(c(1, 1, 2, 3), 2, 2, byrow = TRUE))
layout.show(3)
plot(dose, drugA, type = "l", lty = 1, lwd = 1, cex = 1, col = "red")
plot(dose, drugA, type = "l", lty = 2, lwd = 2, cex = 2, col = "blue")
plot(dose, drugA, type = "l", lty = 3, lwd = 3, cex = 3, col = "green")
par(mypar)


# par
jpeg("./plot/par.jpeg", height = 800, width = 800, quality = 100)
opar <- par(mfrow = c(2, 2))
plot(dose, drugA)
plot(dose, drugA, type = "l", lty = 1, lwd = 1, cex = 1, col = "red")
plot(dose, drugA, type = "l", lty = 2, lwd = 2, cex = 2, col = "blue")
plot(dose, drugA, type = "l", lty = 3, lwd = 3, cex = 3, col = "green")
par(opar)
dev.off()
```

接下来我们讨论如何用 QQ 图（分位数-分位数图） 检查数据是否接近正态分布。

```r
mydata <- read.csv("./qqplot/weight.csv")
w <- mydata$weight

jpeg("./qqplot/qq1.jpeg", height = 600, width = 600, quality = 100)
qqnorm(w)
qqline(w)
dev.off()

print(head(mtcars))
mpg <- mtcars$mpg

jpeg("./qqplot/qq2.jpeg", height = 600, width = 600, quality = 100)
qqnorm(mpg)
qqline(mpg)
dev.off()
```

由上述结果可以看出，样本散点落在给出直线附近时，可以认为接近正态分布。最后我们来讨论一些有意思的图的绘制，比如星型图和脸谱图。
