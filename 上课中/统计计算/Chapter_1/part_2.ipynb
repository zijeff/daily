R 语言之所以能作为统计计算的利器，是因为它可以比较方便地生成特定随机分布，比如正态分布。以下代码都属于 R 里正态分布相关函数的用法。

R 对很多概率分布都提供了统一命名规则：

d*：density，密度函数

p*：distribution，分布函数

q*：quantile，分位数函数

r*：random，随机抽样

值得注意的是，R语言中的正态分布相关函数的参数填入是均值和标准差，而非是一般课本上的均值与方差。

````r
dnorm(4, 2, sqrt(4))    # 在某一点的概率密度
pnorm(4, 2, sqrt(4))    # 在某一点对应的概率分布值

alpha <- 0.05
qnorm(1 - alpha / 2, 0, 1)  # 求对应概率的分位点

rnorm(20, 0, 1)       # 从正态分布中随机取样，这里就生成20个了。
````

讨论完矩阵后，接下来我们依次讨论数据框，array 和 list。我们着重说一下 array 和 list 的区别，array 是同一种数据类型构成的多维数据结构，它可以有很多个维度；而 list 是可以存放不同类型对象的容器，它和向量不同，list 里的每个元素都可以完全不一样。

````r
# data.frame 相关的代码
df <- data.frame(
  Name = c("Alice", "Becka", "James", "Jeffrey", "John"),
  Sex = c("F", "F", "M", "M", "M"),
  Age = c(13, 13, 12, 13, 12),
  Height = c(56.5, 65.3, 57.3, 62.5, 59.0),
  Weight = c(84.0, 98.0, 83.0, 84.0, 99.5)
)
print(df)

print(df$Weight)
print(df$Age)
````

````r
# array 和 list
x <- array(1:12, dim = c(2, 3, 2))
print(x)
print(x[1, , ])

x <- list(1:3, "hello", TRUE)
print(x)
````

有的时候，我们并不希望表达式直接计算出结果，而是希望能直接把表达式存入，可以做到随调随用。要实现这一功能，我们就需要引入 R 语言的 expression 了。

````r
exp1 <- expression(x / (y + exp(z)))    # 构建表达式，不求值
print(exp1)
x <- 3; y <- 2.5; z <- 1    # 给变量赋值
print(eval(exp1))   # 求值
````

逻辑运算具体可以列表如下：
| 运算符    | 含义           | 示例                                   | 结果      |
| ------ | ------------ | ------------------------------------ | ------- |

值得注意的是，R 语言实现了对异或运算的支持，具体如下。


````r
print(xor(TRUE, FALSE))
````
