```r
# 循环，用斐波那契数列作为例子
# for循环
Fib <- rep(0, 20)
Fib[2] <- 1
for (i in 3:20) {
    Fib[i] <- Fib[i - 1] + Fib[i - 2]
}
print(Fib)
# while循环
Fib <- rep(0, 20)
Fib[2] <- 1
i <- 3
while (i <= 20) {
    Fib[i] <- Fib[i - 1] + Fib[i - 2]
    i <- i + 1
}
print(Fib)
# repeat循环
Fib <- rep(0, 20)
Fib[2] <- 1
i <- 3
repeat {
    Fib[i] <- Fib[i - 1] + Fib[i - 2]
    i <- i + 1
    if (i > 20) break
}
print(Fib)
```

```r
# 条件分支执行
x <- 3
if (x > 2) {
    y <- 2 * x
} else {
    y <- 3 * x
}
print(y)
```

R 中的循环与条件控制介绍完了，如果你拥有 C/C++ 基础的话，这里是很好理解的。最后我们在快速地介绍一下函数的定义与使用，以求解整数的阶乘为例子。

```r
# 自然数阶乘
ft <- function(m) 
{
    if (m == 1) {
        res <- 1
    } else {
        res <- m * ft(m - 1)
    }
    return(res)
}
# 看的出来，R 中也可以使用递归调用
print(ft(1))
print(ft(4))
```
